```python
import os
import shutil

# Path where original videos are stored
video_folder = "yoga_videos/Yoga_Vid_Collected"
organized_folder = "organized_yoga_videos"

# Create a new folder for organized videos
os.makedirs(organized_folder, exist_ok=True)

# Go through each file
for filename in os.listdir(video_folder):
    if filename.endswith(".mp4"):
        try:
            # Extract pose name from filename (after "_")
            pose_name = filename.split("_")[1].replace(".mp4", "")
            pose_folder = os.path.join(organized_folder, pose_name)

            # Create folder if it doesn't exist
            os.makedirs(pose_folder, exist_ok=True)

            # Move the video to its pose folder
            src = os.path.join(video_folder, filename)
            dst = os.path.join(pose_folder, filename)
            shutil.copy(src, dst)  # use copy to keep original files safe
        except Exception as e:
            print("Skipping file:", filename, "Error:", e)
```

```python
import pandas as pd

# Folder where videos are now organized
organized_folder = "organized_yoga_videos"

# Prepare a list to hold metadata
metadata = []

# Loop through each folder (pose name)
for pose in os.listdir(organized_folder):
    pose_path = os.path.join(organized_folder, pose)

    # Loop through each file in the pose folder
    for file in os.listdir(pose_path):
        if file.endswith(".mp4"):
            metadata.append({"filename": file, "label": pose})

# Create DataFrame
df = pd.DataFrame(metadata)

# Save to CSV
df.to_csv("organized_yoga_metadata.csv", index=False)

# Show first 5 rows
df.head()
```
